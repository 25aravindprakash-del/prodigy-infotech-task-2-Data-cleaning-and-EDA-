# RMS Titanic Passenger Registry: Demographics & Survival Analysis

This ledger provides a demographic breakdown and survival analysis of the passengers aboard the RMS Titanic during its voyage in April 1912. The analysis covers log reconciliation (data cleaning) and statistical modeling of survival records using a customized maritime green and crimson layout.

**Registry Scope:** Analysis of passenger logs, cabin locations, class structures, and demographic trends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Establish custom visualization palette
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_theme(style="whitegrid")

COLOR_RED = '#9e2a2b'   # Deceased status marker
COLOR_GREEN = '#2b5c3f' # Survived status marker
COLOR_WHITE = '#ffffff'
COLOR_TEXT = '#2d2621'

print("Maritime registry palette initialized.")
```

## 1. Registry Ledger Loading

```python
df = pd.read_csv('train.csv')
df.head()
```

## 2. Record Reconciliation
Analyzing unresolved records to restore incomplete passenger log fields.

```python
print("Unresolved records count per field:")
print(df.isnull().sum())
```

```python
# Reconciling missing age metrics with the median age
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

# Reconciling missing embarkation records with the mode (Southampton)
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_embarked)

# Structural deck assignment based on cabin records
df['Cabin'] = df['Cabin'].fillna('Unknown')
df['Deck'] = df['Cabin'].apply(lambda x: x[0] if x != 'Unknown' else 'Unknown')

print("Reconciled records status:")
print(df.isnull().sum())
```

## 3. Profile Formulation

```python
# Calculate family size on board
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df.head()
```

## 4. Historical Demographics & Survival Analysis
Evaluating demographic distribution variables against passenger status.

### A. Overall Status Breakdown

```python
plt.figure(figsize=(6, 6))
df['Survived'].value_counts().plot.pie(
    labels=['Deceased', 'Survived'], 
    autopct='%1.1f%%', 
    colors=[COLOR_RED, COLOR_GREEN], 
    textprops={'fontsize': 14, 'color': COLOR_TEXT}
)
plt.title('Passenger Status Breakdown', fontsize=16, fontweight='bold', color=COLOR_TEXT)
plt.ylabel('')
plt.show()
```

### B. Status by Gender

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, palette=[COLOR_RED, COLOR_GREEN], errorbar=None)
plt.title('Survival Rate by Gender', fontsize=16, fontweight='bold', color=COLOR_TEXT)
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

### C. Status by Boarding Class

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, palette=[COLOR_GREEN, '#cfc2a9', COLOR_RED], errorbar=None)
plt.title('Survival Rate by Boarding Class', fontsize=16, fontweight='bold', color=COLOR_TEXT)
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

### D. Age Distribution by Status

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Survived'] == 0]['Age'], fill=True, color=COLOR_RED, label='Deceased', alpha=0.6)
sns.kdeplot(df[df['Survived'] == 1]['Age'], fill=True, color=COLOR_GREEN, label='Survived', alpha=0.6)
plt.title('Age Density Profiles by Survival Status', fontsize=16, fontweight='bold', color=COLOR_TEXT)
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### E. Feature Correlation Matrix

```python
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=[np.number])
cmap = sns.diverging_palette(10, 130, as_cmap=True)
sns.heatmap(numeric_df.corr(), annot=True, cmap=cmap, fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix of Ledger Variables', fontsize=16, fontweight='bold', color=COLOR_TEXT)
plt.show()
```

## 5. Summary Findings
1. **Gender Bias**: Female passengers exhibited a significantly higher rate of survival (approx. 74%) compared to male passengers (approx. 19%).
2. **Socio-Economic Factors**: Boarding class structures showed a strong correlation with survival, with 1st Class passengers having high rates of rescue.
3. **Age Influence**: Evacuation patterns favored younger travelers, with higher survival densities among children.
